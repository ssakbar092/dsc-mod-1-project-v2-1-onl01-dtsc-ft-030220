## "tn.movie_budgets.csv.gz" Dataset

```python
cd zippedData/
```

```python
#Imported movie budget dataset
movie_budgets=pd.read_csv("tn.movie_budgets.csv.gz")
movie_budgets
```

```python
# Replacing all the '$' and ',' with a basic integer value

movie_budgets['production_budget'] = movie_budgets['production_budget'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('int64')
movie_budgets['domestic_gross'] = movie_budgets['domestic_gross'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('int64')
movie_budgets['worldwide_gross'] = movie_budgets['worldwide_gross'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('int64')

```

```python
# Creating a new DataFrame to join the tables for later data analysis
# Added a new column NET PROFIT for easier analysis

new_moneybudgets=pd.DataFrame(movie_budgets['movie'])
new_moneybudgets['Movie_Budget']= movie_budgets['production_budget']
new_moneybudgets['National_Profit']=movie_budgets['domestic_gross']
new_moneybudgets['WorldWide_Profit']=movie_budgets['worldwide_gross']
new_moneybudgets['Net_Profit']=movie_budgets['worldwide_gross']-movie_budgets['production_budget']
```

## Joining 2 Datasets "tn.movie_budgets.csv.gz" with "imdb.title.basics.csv.gz" Dataset

```python
cc_df= pd.merge(new_moneybudgets,df2,how='inner',on="movie")
cc_df=cc_df.rename(columns={"start_year":"Year"})
cc_df
```

```python
# Sorting the columns to make sure the dataset is clean
# Rearranging the columns to for easy access
# Applying title method to all the column name
# Dropping all the duplicate values after joining

cc_df=cc_df.drop('original_title',axis=1)
cc_df=cc_df[['movie','genres','Movie_Budget','National_Profit','WorldWide_Profit','Net_Profit','Year','runtime_minutes']]
cc_df.columns = map(str.title, cc_df.columns)
cc_df.drop_duplicates(subset='Movie',keep="first")
cc_df
```

```python
# Defining a function to count an individual genre in the dataset

def counter(df,col):
    genre_dict={}
    for genres in df[col]:
        for genre in genres.split(","):
            if genre in genre_dict:
                genre_dict[genre]+=1
            else:
                genre_dict[genre]=1
    return genre_dict
```

```python
totalcount=counter(cc_df,'Genres')
print(totalcount)
```

```python
# Converting the dictionaries int0 a list to  plot a graph and count the number of movies in particular genre
x=list(totalcount.keys())
y=list(totalcount.values())
```

```python
# Plotting a graph in relation to the number of movies with the respected genres

plt.figure(figsize=(25,10),dpi=100)
plt.style.use('fivethirtyeight')
plt.bar(x,y,label='Different Genres',color='#556b2f')
plt.plot(x,y,label='Number of Movies',color='#0000A0')
plt.title("Genre of Top 100 movies", fontsize=25)
plt.xlabel('Genres',fontsize=35)
plt.ylabel('Number of movies',fontsize=35)
plt.legend(loc=1)
plt.show()
```
